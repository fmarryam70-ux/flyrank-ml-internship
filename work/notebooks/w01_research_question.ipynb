## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Unit of analysis: one content page, per client, over a defined time window.
Decision: which pages should a content reviewer look at first this week?
Action: a reviewer opens the top-ranked pages and decides to refresh, expand,
or leave them.
Cost of a wrong call: if a low-priority page is ranked too high, the reviewer
wastes limited time on a page that didn't need attention. If a genuinely
declining page is ranked too low, it keeps losing visibility unnoticed. Both
costs are about wasted reviewer capacity, not catastrophic — but capacity is
limited, so ranking quality matters.
Why ML helps: with tens of thousands of pages and only a few reviewers, a
learned model can rank pages by real risk/opportunity far faster and more
consistently than manual spot-checking.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What this work CAN say:
- Which pages show observed patterns of decline, staleness, or low CTR relative to their peers
- A ranked, evidence-backed list of pages a reviewer should look at first
- Directional signals: "this page's engagement is below what similar pages typically get"

What this work CANNOT say:
- That refreshing a page will cause it to recover (that needs a real experiment, not this data)
- That any signal here reflects Google's actual ranking algorithm
- That a page is "declining" for certain — only that it shows patterns associated with decline;
  consolidation, seasonality, or noise could explain the same pattern
- Any claim beyond "observed" and "directional" — this is decision support for a human reviewer,
  not an automated verdict
